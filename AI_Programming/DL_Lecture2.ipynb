# 1. Linear Regression

```python
import numpy as np

x = np.array([1, 3, 5, 7, 9])
y = np.array([75, 77, 85, 83, 90])
mx = np.mean(x)
my = np.mean(y)
print('mean of x: {}, mean of y: {}'.format(mx, my))

down = sum([(i - mx)**2 for i in x]) #sum([a, b, c])

top = 0
for i in range(len(x)):
    top = top + (x[i] - mx) * (y[i] - my)
a = top / down
b = my - (mx * a)

print('regression line: y = {}x + {}'.format(a, b))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([1, 3, 5, 7, 9])
x2 = np.array([0, 1, 5, 2, 4])
y = np.array([75, 77, 85, 83, 90])

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x1, x2, y)
plt.title("Scatter plot")
plt.show()

a1 = 0
a2 = 0
b = 0
lr = 0.0003 # learning rate
epochs = 20001 # epochs
n = len(x1) # number of x

# gradient decent
for i in range(epochs):
  y_pred = a1 * x1 + a2 * x2 + b
  error = y - y_pred
  a1_diff = (2/n) * sum(-x1 * error)
  a2_diff = (2/n) * sum(-x2 * error)
  b_diff = (2/n) * sum(-error)

  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  if(i % 2000 == 0):
    print('epoch: {}, a1: {}, a2: {}, b: {}'.format(i, a1, a2, b))

print('a1_diff : {}, a2_diff : {}, b_diff : {}'.format(a1_diff, a2_diff, b_diff))
y_pred = a1 * x1 + a2 * x2 + b# regression with the last value of a and b.
print('y:', y)
print('y_pred:', y_pred)
```

# 2. Multivariate Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([1, 3, 5, 7, 9])
y = np.array([75, 77, 85, 83, 90])

model = Sequential()
model.add(Dense(1, input_dim = 1, activation = 'linear'))
model.compile(optimizer='sgd', loss='mse')
model.fit(x, y, epochs=2000)

y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()

hours = 7
exp_score = model.predict([hours])
print("The expected score is", exp_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[1, 0], [3, 1], [5, 5], [7, 2], [9, 4]]) # input dimension: 2
y = np.array([75, 77, 85, 83, 90])

model = Sequential()
model.add(Dense(1, input_dim = 2, activation = 'linear'))
model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x, y, epochs = 2000)
y_pred = model.predict(x)

hours = 6
genealogy = 4
exp_score = model.predict([[hours, genealogy]])
print('The expected score is', exp_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

x = np.array([1, 3, 5, 7, 9])
y = np.array([75, 77, 85, 83, 90])
x = x.astype(np.float32)
y = y.astype(np.float32)

# User-defined squared 2-norm loss function
def squared_2norm_loss(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))

model = Sequential()
model.add(Dense(1, input_dim=1, activation='linear'))
model.compile(optimizer='sgd', loss=squared_2norm_loss)
model.fit(x, y, epochs=2000, verbose=1)
y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()

hours = 7
exp_score = model.predict([hours])
print('The expected score is', exp_score)
```

# 3. Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([1, 3, 5, 7, 9, 11, 13])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()
model.add(Dense(1, input_dim = 1, activation = 'sigmoid'))

model.compile(optimizer = 'sgd', loss = 'binary_crossentropy')
model.fit(x, y, epochs = 10000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

#inference
hours = 6
prediction = model.predict([hours])
print('{}시간 공부하면 합격 확률은 {}입니다.'.format(hours, prediction))
```

# Practice

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

x = np.array([x for x in range(1, 1001)])
y = np.array([1 if i>=500 else 0 for i in x])

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(64, activation='linear', name='Dense_1'))
model.add(Dense(32, activation='linear', name='Dense_2'))
model.add(Dense(16, activation='linear', name='Dense_3'))
model.add(Dense(8, activation='linear', name='Dense_4'))
model.add(Dense(1, activation='sigmoid', name='Dense_5'))
model.summary()

model.compile(optimizer='adam', loss='binary_crossentropy')
model.fit(x, y, epochs=10000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

hours = 6
prediction = model.predict([hours])
print('{}시간 공부하면 합격 확률은 {}입니다.'.format(hours, prediction))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import to_categorical

x = np.array([x for x in range(1, 1001)])
y = np.array([1 if i>=500 else 0 for i in x])

y_categorical = to_categorical(y, num_classes=2)

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(64, activation='linear', name='Dense_1'))
model.add(Dense(32, activation='linear', name='Dense_2'))
model.add(Dense(16, activation='linear', name='Dense_3'))
model.add(Dense(8, activation='linear', name='Dense_4'))
model.add(Dense(2, activation='softmax', name='Dense_5'))
model.summary()

model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit(x, y_categorical, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

hours = 510
prediction = model.predict([hours])
print('{}시간 공부하면 합격 확률은 {}입니다.'.format(hours, prediction))
```
