```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3, 35004]

plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')
# plt.plot(years, gdp)
plt.title("GDP per capita")
plt.xlabel("years")
plt.ylabel("dollars")
plt.savefig("gdp_per_capita.png", dpi = 600)
plt.show()
```

```python
plt.plot(years, gdp, color = 'red', marker = 'x', markersize = 12, linestyle = 'dashed')
plt.title("GDP per capita")
plt.xlabel("years")
plt.ylabel("dollars")
plt.savefig("gdp_per_capita.png", dpi = 600)
plt.show()
```

```python
x = [x for x in range(-20, 20)]
y1 = [2*i for i in x]
y2 = [i**2 - 4*i + 3 for i in x]
y3 = [-i**2 - 4*i - 3 for i in x]

plt.title("Three graphs together")
#plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.plot(x, y1, 'r--', label = 'y=2x')
plt.plot(x, y2, 'g^-', label = 'y=x^2-4x+3')
plt.plot(x, y3, 'b*:', label = 'y=-x^2-4x-3')

plt.axis([-30, 30, -40, 40])
plt.legend()
plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3, 35004]

plt.title("Bar exercise")
plt.bar(range(len(years)), gdp, label = 'gdp per capita')
plt.legend()
plt.xlabel("Year")
plt.ylabel("USD")
plt.xticks(range(len(years)), years)
plt.show()
```

```python
x = np.arange(0, 30)
y = x + np.random.randn(30)

plt.title('Scatter plot exercise')
plt.scatter(x, y) #plt.plot(x, y)
plt.xlabel("time")
plt.ylabel("random")
plt.show()
```

```python
lgtwins = [8, 9, 2, 2, 3, 4, 7, 2, 4, 3, 7, 1, 3, 3, 5, 10, 2, 2, 1]

plt.hist(lgtwins, bins = 10, color = 'red')
plt.title('LG Twins standings')
plt.xlabel('Years')
plt.ylabel('Standings')
plt.show()
```

```python
n_bins = 20
gauss1 = np.random.randn(1000)
gauss2 = np.random.randn(1000) + 3

plt.title('Two Normal distribution')
plt.hist(gauss1, n_bins, color = 'blue', alpha = 0.5)
plt.hist(gauss2, n_bins, color = 'red', alpha = 0.5)
plt.show()
```
