```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.layers import Embedding, Dense, GlobalAveragePooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
# 텍스트 데이터 (긍정/부정 예시)
texts = ["이 영화 정말 최고에요! 너무 감동적이에요.",
         "정말 재미있고 흥미진진했어요!",
         "다시 보고 싶은 영화에요.",
         "너무 지루하고 기대 이하였어요.",
         "이건 정말 시간 낭비였어요. 완전 별로에요.",
         "돈이 아까운 영화였어요."
]
# 레이블 (긍정: 1, 부정: 0)
labels = [1, 1, 1, 0, 0, 0]
```

```python
# 토크나이저 설정 및 텍스트 시퀀스 변환
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# 패딩 추가 (최대 시퀀스 길이에 맞춰 0으로 패딩)
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')
```

```python
# 워드 인덱스의 수
vocab_size = len(tokenizer.word_index) + 1
# 모델 생성
embedding_dim = 8 # 임베딩 벡터 차원
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len),
    GlobalAveragePooling1D(), # 간단한 풀링 사용 가능
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid') # 이진 분류이므로 sigmoid 사용
])
```

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(padded_sequences, np.array(labels), epochs=100, verbose=1)
```

```python
# 새로운 문장 테스트
# test_text = "정말 멋진 영화였어요. 다시 보고 싶어요."
test_text = "추천하지 않습니다."
test_sequence = tokenizer.texts_to_sequences([test_text])
padded_test_sequence = pad_sequences(test_sequence, maxlen=max_len, padding='post')

# 예측 수행
prediction = model.predict(padded_test_sequence)
print(f"Prediction (긍정 확률): {prediction[0][0]:.4f}")

# 예측 결과 해석
if prediction[0][0] >= 0.5:
    print("긍정적인 감정으로 예측되었습니다.")
else:
    print("부정적인 감정으로 예측되었습니다.")
```
