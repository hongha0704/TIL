```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([1, 3, 5, 7, 9, 11, 13])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()
#model.add(Dense(10, input_dim=1, activation='relu')) # 은닉층 추가
model.add(Dense(1, input_dim = 1, activation = 'sigmoid'))

model.compile(optimizer = 'sgd', loss = 'binary_crossentropy')
model.fit(x, y, epochs = 2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

#inference
hours = 8
prediction = model.predict(np.array([hours]))
print('{}시간 공부하면 합격 확률은 {}입니다.'.format(hours, prediction))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

x = np.array([1, 3, 5, 7, 9, 11, 13])
y = np.array([0, 0, 0, 1, 1, 1, 1])

y_categorical = to_categorical(y, num_classes=2)

model = Sequential()
model.add(Dense(10, input_dim=1, activation='relu')) # 은닉층 추가
model.add(Dense(2, activation='softmax')) # 다중 클래스 분류, 2개의 출력 노드

model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x, y_categorical, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

# inference
hours = 8
prediction = model.predict(np.array([hours]))
print('{}시간 공부하면 합격 확률은 {}입니다.'.format(hours, prediction))
```
