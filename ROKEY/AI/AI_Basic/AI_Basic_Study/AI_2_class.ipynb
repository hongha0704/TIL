```python
def sigmoid(x, a):
    return 1/(1 + np.exp(-a*x))

xp = np.linspace(-3, 3, 61)
yp = sigmoid(xp, 1.0)
yp2 = sigmoid(xp, 2.0)
```

```python
plt.plot(xp, yp)
plt.plot(xp, yp2)

plt.show()
```

```python
plt.plot(xp, yp, label='sigmoid 1', lw=3, c='k')
plt.plot(xp, yp2, label='sigmoid 2', lw=2, c='b')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

image = mnist.data
label = mnist.target
```

```python
plt.figure(figsize=(10, 3))

for i in range(20):
    ax = plt.subplot(2, 10, i+1)
    img = image[i].reshape(28, 28)
    ax.imshow(img, cmap='gray_r')
    ax.set_title(label[i])

    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
