```python
class Calculator:
  def __init__(self, a:int, b:int):
    self.a = a
    self.b = b

  def add(self):
    return self.a + self.b

  def sub(self):
    return self.a + self.b
```

```python
calculator = Calculator(10, 20)
```

```python
print(calculator.add())
print(calculator.sub())
```

```python
class Calculator2(Calculator):
  def __init__(self, a:int, b:int, r:int):
    super().__init__(a, b)
    self.r = r

  def mul(self):
    return self.a * self.b

  def div(self):
    return self.a / self.b

calculator2 = Calculator2(10, 20, 2)
```

```python
print(calculator2.add())
print(calculator2.sub())
print(calculator2.mul())
print(calculator2.div())
```

```python
import numpy as np

x = np.arange(16).reshape(4, -1)
print(x)
```

```python
## Sigmoid graph

def sigmoid(x, w=1):
  y = 1 / (1 + np.exp(-w*x))
  return y
```

```python
x = np.linspace(-5, 5, 100)
print(x)
```

```python
y = sigmoid(x)
y1 = sigmoid(x, 0.5)
y2 = sigmoid(x, 2)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, label='w=1')
plt.plot(x, y1, label='w=0.5')
plt.plot(x, y2, label='w=2')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784", version=1, as_frame=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = mnist.data
print(image.shape)

label = mnist.target
print(label)
```

```python
idx = np.random.randint(0, 70000)
print("idx =", idx)
print("target =", label[idx])
sample = image[idx].reshape(28, 28)

plt.imshow(sample, cmap= 'gray')
plt.show()
```

```python
plt.figure()

for i in range(20):
  ax = plt.subplot(2, 10, i+1)
  img = image[i].reshape(28, 28)
  ax.imshow(img, cmap= 'gray')
  ax.set_title(label[i])
  ax.axis("off")

plt.show()
```
