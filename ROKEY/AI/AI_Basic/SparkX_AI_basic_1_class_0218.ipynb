```python
## sum function
def addNum(num1:int, num2:int = 0) -> float:
  sum = num1 + num2
  return float(sum)

result1 = addNum(10, 2)
print("result1 =", result1)

result2 = addNum(10)
print("result2 =", result2)
```

```python
def add_sub(num1, num2):
    add = num1 + num2
    sub = num1 - num2
    return add, sub

add_sub(1, 2) # tuple로 반환
x = add_sub(3, 4)  # tuple로 반환
print('x = ', x)

add, sub = add_sub(3, 4)
print("add = {}, sub = {}".format(add, sub))
```

```python
## 함수내 함수호출, stack 구조
def mul(a, b):
    return a*b

def divide(a, b):
    return a/b

def mul_div(x, y):
    num1 = mul(x, y)
    num2 = divide(x, y)
    return num1, num2

result = mul_div(2, 1)
print('result = ', result)
```

```python
## class 만들기 : 비슷한 기능들을 한 번에 관리
class Calculator:
    def __init__(self, a:int, b:int):
        self.result = 0
        self.a = a
        self.b = b

    def add(self):
        self.result = self.a + self.b
        return self.result

    def sub(self):
        self.result = self.a - self.b
        return self.result
```

```python
calculator = Calculator(3, 2)
print("add = ", calculator.add())
print("sub = ", calculator.sub())
```

```python
## 상속
class CalculatorV2(Calculator):
    def __init__(self, x, y, c):
        super().__init__(x, y)
        self.c = c

    def mul(self):
        self.result = self.a * self.b
        return self.result

    def div(self):
        self.result = self.a / self.b
        return self.result

    # def __str__(self):
    #     return str(self.c)
```

```python
## 상속
class Calculator2(Calculator):
    def __init__(self, x, y):
        super().__init__(x, y)
        self.s = 'Calculator2 from calculator1'

    def mul(self):
        self.result = self.a * self.b
        return self.result

    def div(self):
        self.result = self.a / self.b
        return self.result

    def __str__(self):
        return self.s
```

```python
calculator2 = Calculator2(3, 2)

print("add = ", calculator2.add())
print("sub = ", calculator2.sub())
print("mul = ", calculator2.mul())
print("div = ", calculator2.div())
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def draw(self):
    plt.plot(self.x, self.y, marker="o", markersize=10, c="k")
```

```python
p1 = Point(2, 3)
p2 = Point(-1, -2)

p1.draw()
p2.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

```python
class Circle1(Point):
  def __init__(self, x, y, r):
    super().__init__(x, y)
    self.r = r

c1_1 = Circle1(1, 0, 2)

## attribute
print(c1_1.x)
print(c1_1.y)
print(c1_1.r)
```

```python
p1.draw()
p2.draw()
c1_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```
