```python
## w1x1 + w2x2 + w3x3 + ... + b
def perceptron(x, w, b=0):
  if not isinstance(x, np.ndarray) or not isinstance(w, np.ndarray):
    x = np.array(x)
    w = np.array(w)
  y = np.sum(w*x) + b # element-wise mul 곱셈
  return y
```

```python
x = np.array([1, 2, 3])
w = np.array([1, 2, 3])

y = perceptron(x, w)
print("result =", y)
```

```python
## random initialize
# np.manual_seed(221)
def perceptron(x):
  w = np.random.rand(len(x))
  b = np.random.rand()
  y = np.sum(w*x) + b
  return y, w, b
```

```python
x = np.array([1, 3, 5, 7, 9])

y, w, b = perceptron(x)
print("w =", w)
print("b =", b)
print("y =", y)
```

```python
## Forward propagation, 순전파
np.random.seed(221)

def perceptron(x):
  w = np.random.rand(len(x))
  b = np.random.rand()
  y = np.sum(w*x) + b
  return y, w, b

def sigmoid(x):
  y = 1 / (1+np.exp(-x))
  return y

x = np.random.rand(5)
print(x)

##
y, w, b = perceptron(x)
print("y =\n", y)

p = sigmoid(y)
print("p =\n", p)
```
