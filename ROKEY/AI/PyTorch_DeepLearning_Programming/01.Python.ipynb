```python
def f(x):
    return (2 * x**2 + 2)
```

```python
x = np.arange(-2, 2.1, 0.25)
print(x)
```

```python
y = f(x)
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.show()
```

```python
def f1(x):
    return (x**2)

def f2(x):
    return (x*2)

def f3(x):
    return (x+2)

x1 = f1(x)
x2 = f2(x1)
y = f3(x2)

print(y)

plt.plot(x, y)
plt.show()
```

```python
def fdiff(f):
    def diff(x):
        h = 1e-6
        return (f(x+h) - f(x-h)) / (2*h)
    return diff
```

```python
diff = fdiff(f)

y_dash = diff(x)

print(y_dash)
```

```python
plt.plot(x, y, label=r'y = f(x)', c='b')
plt.plot(x, y_dash, label=r"y = f'(x)", c='k')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def draw(self):
        plt.plot(self.x, self.y, marker='o', markersize=10, c='k')
```

```python
p1 = Point(2, 3)
p2 = Point(-1, -2)
```

```python
print(p1.x, p1.y)
print(p2.x, p2.y)
```

```python
p1.draw()
p2.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.show()
```

```python
class Circle1(Point):
    def __init__(self, x, y, r):
        super().__init__(x, y)
        self.r = r
```

```python
c1_1 = Circle1(1, 0, 2)
print(c1_1.x, c1_1.y, c1_1.r)
```

```python
ax = plt.subplot()
p1.draw()
p2.draw()
c1_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.show()
```

```python
class Circle2(Point):
    def __init__(self, x, y, r):
        super().__init__(x, y)
        self.r = r

    def draw(self):
        c = patches.Circle(xy=(self.x, self.y), radius=self.r, fc='b', ec='k')
        ax.add_patch(c)
```

```python
c2_1 = Circle2(1, 0, 2)
print(c2_1.x, c2_1.y, c2_1.r)
```

```python
ax = plt.subplot()
p1.draw()
p2.draw()
c2_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.show()
```

```python
class Circle3(Point):
    def __init__(self, x, y, r):
        super().__init__(x, y)
        self.r = r

    def draw(self):
        super().draw()
        c = patches.Circle(xy=(self.x, self.y), radius=self.r, fc='b', ec='k')
        ax.add_patch(c)
```

```python
c3_1 = Circle3(1, 0, 2)
print(c3_1.x, c3_1.y, c3_1.r)
```

```python
ax = plt.subplot()
p1.draw()
p2.draw()
c3_1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.show()
```

```python
class H:
    def __call__(self, x):
        return 2*x**2 + 2
```
